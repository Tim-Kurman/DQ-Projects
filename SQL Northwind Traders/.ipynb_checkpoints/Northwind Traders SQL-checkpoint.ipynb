```python
%%sql
SELECT 
    o.orderdate AS order_date,
    p.productname AS product,
    od.quantity AS quantity,
    p.price AS price,
    od.quantity * p.price AS USD_sales,
    cat.categoryname AS category,
    c.customername AS customer,
    c.country AS customer_country,
    e.firstname || ' ' || e.lastname AS employee
FROM orders o
JOIN orderdetails od ON o.orderid = od.orderid
JOIN products p ON od.productid = p.productid
JOIN categories cat ON p.categoryid = cat.categoryid
JOIN customers c ON o.customerid = c.customerid
JOIN employees e ON o.employeeid = e.employeeid
LIMIT 10;
```

```python
%%sql
SELECT e.employeeid, e.firstname || ' ' || e.lastname AS Employee,
           SUM(p.price * od.quantity) AS "Total Sales"
FROM orders o
JOIN orderdetails od ON o.orderid = od.orderid
JOIN employees e ON o.employeeid = e.employeeid
JOIN products p ON od.productid = p.productid

GROUP BY e.employeeid;
```

```python
%%sql
WITH EmployeeSales AS (
    SELECT e.employeeid, e.firstname || ' ' || e.lastname AS Employee,
           SUM(p.price * od.quantity) AS "Total Sales"
    FROM orders o
    JOIN orderdetails od ON o.orderid = od.orderid
    JOIN employees e ON o.employeeid = e.employeeid
    JOIN products p ON od.productid = p.productid

    GROUP BY e.employeeid
)
SELECT employeeid, Employee,
       RANK() OVER (ORDER BY "Total Sales" DESC) AS "Sales Rank"
FROM EmployeeSales;
```

```python
%%sql
SELECT 
    DATE_TRUNC('month', orderdate)::DATE AS month, 
    SUM(p.price * od.quantity) AS total_sales
FROM orders o
JOIN orderdetails od ON o.orderid = od.orderid
JOIN products p ON od.productid = p.productid
GROUP BY DATE_TRUNC('month', orderdate)
ORDER BY month;
```

```python
%%sql
WITH Monthly_Sales AS (
    SELECT 
        DATE_TRUNC('month', orderdate)::DATE AS month, 
        SUM(p.price * od.quantity) AS total_sales
    FROM orders o
    JOIN orderdetails od ON o.orderid = od.orderid
    JOIN products p ON od.productid = p.productid
    GROUP BY DATE_TRUNC('month', orderdate)
    )
SELECT month,
       SUM(total_sales) OVER (ORDER BY month) AS "Running Total"
FROM Monthly_Sales
ORDER BY Month;
```

```python
%%sql
WITH Monthly_Sales AS (
    SELECT 
        EXTRACT(month FROM orderdate) AS month,
        EXTRACT(year FROM orderdate) AS year,
        SUM(p.price * od.quantity) AS total_sales
    FROM orders o
    JOIN orderdetails od ON o.orderid = od.orderid
    JOIN products p ON od.productid = p.productid
    GROUP BY year, month
    )
SELECT 
    month, 
    year, 
    total_sales, 
    LAG(total_sales) OVER (ORDER BY year, month) AS PreviousMonthSales
FROM Monthly_Sales;
```

```python
%%sql
WITH 
Monthly_Sales AS (
    SELECT 
        EXTRACT(month FROM orderdate) AS month,
        EXTRACT(year FROM orderdate) AS year,
        SUM(p.price * od.quantity) AS total_sales
    FROM orders o
    JOIN orderdetails od ON o.orderid = od.orderid
    JOIN products p ON od.productid = p.productid
    GROUP BY year, month
    ),
Lagged_Sales AS(
    SELECT
        month, 
        year, 
        total_sales, 
        LAG(total_sales) OVER (ORDER BY year, month) AS PreviousMonthSales
    FROM Monthly_Sales
    )
SELECT year, month,
       ((total_sales - previousmonthsales)/previousmonthsales) * 100 AS "Growth Rate"
FROM Lagged_Sales;
```

```python
%%sql
SELECT
    o.orderid,
    o.customerid,
    SUM(p.price * od.quantity) AS ordervalue
FROM orders o
JOIN orderdetails od ON o.orderid = od.orderid
JOIN products p ON od.productid = p.productid
GROUP BY o.customerid, o.orderid
LIMIT 10;
```

```python
%%sql
WITH OrderValue AS
    (
    SELECT
        o.orderid,
        o.customerid,
        SUM(p.price * od.quantity) AS ordervalue
    FROM orders o
    JOIN orderdetails od ON o.orderid = od.orderid
    JOIN products p ON od.productid = p.productid
    GROUP BY o.customerid, o.orderid
    )
SELECT
    customerid,
    orderid,
    ordervalue,
    CASE
        WHEN ordervalue > AVG(ordervalue) OVER () THEN 'Above Average'
        ELSE 'Below Average'
    END AS "Value Category"
FROM OrderValue
LIMIT 10;
```

```python
%%sql
WITH CategorySales AS (
    SELECT cat.categoryid,
           cat.categoryname,
           SUM(p.price * od.quantity) AS totalsales
    FROM categories cat
    JOIN products p ON cat.categoryid = p.categoryid
    JOIN orderdetails od ON p.productid = od.productid
    GROUP BY cat.categoryid
)
SELECT categoryid, categoryname,
       totalsales / SUM(totalsales) OVER () * 100 AS "Sales Percentage"
FROM CategorySales;
```

```python
%%sql
WITH ProductSales AS (
    SELECT p.categoryid, 
           p.productid, p.productname,
           SUM(p.price * od.quantity) AS totalsales
    FROM products p
    JOIN orderdetails od ON p.productid = od.productid
    GROUP BY p.categoryid, p.productid
)
SELECT categoryid, 
       productid, productname,
       totalsales
FROM (
    SELECT categoryid, 
           productid, productname,
           totalsales, 
           ROW_NUMBER() OVER (PARTITION BY categoryid ORDER BY totalsales DESC) AS rn
    FROM ProductSales
) tmp
WHERE rn <= 3;
```
